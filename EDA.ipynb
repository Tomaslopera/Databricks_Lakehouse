```python
from pyspark.sql.functions import sum, col, isnan
```

## Movies

```python
movies = spark.table("movielens.bronze.movie")
display(movies)
```

```python
nulls = movies.select([sum(col(c).isNull().cast("int")).alias(c) for c in movies.columns])
nulls.show(truncate=False)
```

```python
movies.printSchema()
```
